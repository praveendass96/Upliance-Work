```python
df3.drop(columns=['User ID','Dish Name','Meal Type'],inplace=True)
```

```python
df3.head()
```

```python
session_orders=pd.merge(left=df2,right=df3,on='Session ID',how='inner')
session_orders.head()
```

```python
session_orders.drop(columns=['Session Start','Session End'],inplace=True)
#no need for the columns(session start, session end) we have already extracted the time as durations
```

```python
pd.merge(left=df1,right=session_orders,on='User ID',how='inner').info()
```

```python
md=pd.merge(left=df1,right=session_orders,on='User ID',how='inner')
```

```python
md.head()
```

```python
md.to_csv('merged_data.csv', index=False)
# converted the merged data into excel file
```

```python
md.info()
```

```python
# the merged data has 16 rows and 20 columns there is no loss of data and now we move on to visualisation
#Analysis
```

```python
# relationship between cooking sessions and user orders
md.groupby('User ID')['Total Orders'].count()
```

```python
pd.pivot_table(md, index=["User ID"],
              columns = ["Order Date"],
              values=["Total Orders"])
```

```python
#we can clearly see that there were no orders placed by user 9 and 10
```

```python
md.groupby('User ID')['Total Orders'].aggregate(["sum","min","max","mean"])
# maximum orders placed is by user003 of 45 orders
#minimum orders placed is by user008 of 5 orders
```

```python
md.groupby(['Dish Name', 'Meal Type']).agg({
    'Order ID': 'count',
    'Session Rating': 'mean',
    'Rating': 'mean'
})
# dishes order for Breakfast are Oatmeal and Pancakes
# for lunch is Caesar salad and Veggie Burger
# for dinner is Grilled chicken and spaghetti
```

```python
Relation=md.groupby(['Dish Name', 'Meal Type']).agg({
    'Order ID': 'count',
    'Session Rating': 'mean',
    'Rating': 'mean'
}).rename(columns={
    'Order ID': 'order_count', 
    'Session Rating': 'avg_session_rating', 
    'Rating': 'avg_order_rating'
}).reset_index()
```

```python
Relation.head()
```

```python
# identify popular dishes
PD=Relation.sort_values(by='order_count', ascending=False)
PD.head()
# to identify the popular dishes sort in descending order with order count grilled chicken and spaghetti tops the list
```

```python
# Demographic factors influencing behaviours
DA= md.groupby(['Location', 'Age']).agg({
    'Total Orders': 'sum',
    'Session Rating': 'mean',
    'Amount (USD)': 'mean'
}).reset_index()
DA.head()
# Chicago placed maximum orders of 45 and followed by los angeles 35
# Also from the above inference till now there was no orders placed by user 9(dallas) and 10(phoenix) in the year 2024 but they have 
# total orders of 6 and 8 respectively and hence we can see the data is missing or not updated.
```

```python
# Plot popular dishes
plt.figure(figsize=(10, 6))
sns.barplot(data=PD.head(10), x='order_count', y='Dish Name')
plt.title('Top 10 Popular Dishes')
plt.xlabel('Number of Orders')
plt.ylabel('Dish Name')
plt.tight_layout()
plt.savefig('popular_dishes.png')
```

```python
# Plot demographic factors
plt.figure(figsize=(10, 6))
sns.boxplot(data=md, x='Favorite Meal', y='Amount (USD)', hue='Location')
plt.title('Spending Patterns by Location and Favorite Meal')
plt.xlabel('Favorite Meal')
plt.ylabel('Average Spending (USD)')
plt.legend(title='Location')
plt.tight_layout()
plt.savefig('demographics_analysis.png')
```
